## Handling Missing Values

In this section, missing values in the dataset are identified and handled appropriately.

The following operations are performed:
- Counting missing values in each column
- Calculating total missing values
- Filling missing numerical values using mean
- Filling missing categorical values using placeholder values
- Tracking how many values were filled

```python
# Checking missing values in each column

missing_values_before = df.isnull().sum()

print("Missing Values Before Cleaning:\n")
print(missing_values_before)
```

```python
# Total missing values in dataset

total_missing_before = df.isnull().sum().sum()

print("Total Missing Values Before Cleaning:", total_missing_before)
```

### Dropping Columns with Excessive Missing Values

Some columns contain a very high number of missing values and are not important for this basic data cleaning assignment.

The following columns are removed:
- `videos`
- `what_customers_said`
- `variations`
- `seller_information`

These columns mainly contain multimedia or highly complex nested information.

```python
# Columns to drop

columns_to_drop = [
    'videos',
    'what_customers_said',
    'variations',
    'seller_information'
]

# Dropping columns

df = df.drop(columns=columns_to_drop)

print("Dropped Columns:")
print(columns_to_drop)
```

### Filling Remaining Missing Values

After removing highly incomplete columns, the remaining missing values are handled.

- Numerical values are filled using mean
- Categorical values are filled using 'Unknown'

```python
# Missing values before filling

discount_missing_before = df['discount'].isnull().sum()

seller_missing_before = df['seller_name'].isnull().sum()
```

```python
# Filling missing numerical values

df['discount'] = df['discount'].fillna(df['discount'].mean())

# Filling missing categorical values

df['seller_name'] = df['seller_name'].fillna('Unknown')
```

```python
# Checking missing values after cleaning

remaining_missing = df.isnull().sum()

print("Remaining Missing Values:\n")
print(remaining_missing)
```

```python
# Cleaning summary

print("Data Cleaning Summary\n")

print(f"Total Missing Values Before Cleaning: {total_missing_before}")

print(f"Columns Dropped: {len(columns_to_drop)}")

print(f"Discount Values Filled: {discount_missing_before}")

print(f"Seller Name Values Filled: {seller_missing_before}")

total_filled = discount_missing_before + seller_missing_before

print(f"Total Values Filled: {total_filled}")

remaining_total_missing = df.isnull().sum().sum()

print(f"Remaining Missing Values After Cleaning: {remaining_total_missing}")
```

## Removing Duplicate Records

Duplicate records can affect data quality and analysis results.

In this section:

```python
# Counting duplicate rows

duplicate_count = df.duplicated().sum()

print("Total Duplicate Rows:", duplicate_count)
```

### Removing Duplicate Rows

Duplicate rows are removed using the `drop_duplicates()` function to improve dataset quality.

Note: Here there are no duplicates in this dataset/dataframe but there could be, for those we can use following code snippet to remove duplicates.

```python
# Shape before removing duplicates

shape_before = df.shape

# Removing duplicate rows

df = df.drop_duplicates()

# Shape after removing duplicates

shape_after = df.shape

print("Shape Before Removing Duplicates:", shape_before)

print("Shape After Removing Duplicates:", shape_after)
```

```python
# Duplicate removal summary

rows_removed = shape_before[0] - shape_after[0]

print("Duplicate Removal Summary\n")

print(f"Duplicate Rows Found: {duplicate_count}")

print(f"Rows Removed: {rows_removed}")

print(f"Final Number of Rows: {shape_after[0]}")
```

# Cleaning and Transforming Price Data

The `final_price` column contains currency symbols and commas, which makes it unsuitable for mathematical operations.

In this section:
- Currency symbols are removed
- Commas are removed
- The column is converted into numeric datatype
- Statistical analysis is performed on cleaned prices

```python
# Checking datatype before cleaning

print("Datatype Before Cleaning:")

print(df['final_price'].dtype)
```

## Removing Currency Symbols and Converting to Numeric Format

The `final_price` column currently contains:
- Currency symbols (`₹`)
- Commas (`,`)
- Extra quotation marks (`"`)

These characters are removed before converting the column into float datatype.

```python
# Cleaning final_price column

df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
    .str.strip()
)

# Converting to numeric datatype

df['final_price'] = pd.to_numeric(df['final_price'])

print("Price Column Cleaned Successfully")
```

```python
# Checking datatype after cleaning

print("Datatype After Cleaning:")

print(df['final_price'].dtype)
```

```python
# Displaying cleaned price values

df['final_price'].head()
```

# Creating Derived Columns

The dataset does not contain a `quantity` column.

For demonstration and analysis purposes:
- A sample quantity column is created
- A new derived column `total_amount` is generated using:

`total_amount = final_price × quantity`

```python
# Creating random quantity values between 1 and 5

np.random.seed(42)

df['quantity'] = np.random.randint(1, 6, size=len(df))

# Display quantity column

df[['final_price', 'quantity']].head()
```

## Creating Total Amount Column

A new column named `total_amount` is created using:

`total_amount = final_price × quantity`

```python
# Creating total_amount column

df['total_amount'] = df['final_price'] * df['quantity']

# Displaying new columns

df[['final_price', 'quantity', 'total_amount']].head()
```

```python
# Statistical summary of total amount

df['total_amount'].describe()
```
